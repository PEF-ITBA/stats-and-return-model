```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
ms = pd.read_csv('data/psfa_facebook.csv')
ms.head()
```

```python
ms['logReturn'] = np.log(ms['Close'].shift(-1)) - np.log(ms['Close'])
```

```python
sample_size = ms['logReturn'].shape[0]
sample_mean = ms['logReturn'].mean()
sample_std = ms['logReturn'].std(ddof=1) / sample_size**0.5
```

```python
# Usando distribuci√≥n Z
# left and right quantile
z_left = norm.ppf(0.05)
z_right = norm.ppf(0.95)

# upper and lower bound
interval_left = sample_mean + z_left*sample_std
interval_right = sample_mean + z_right*sample_std
print(f'Hay un 90% de chances de que el promedio del retorno este en el intervalo ({interval_left}, {interval_right}))')
```

```python
# Diretamente
interval_left = norm.ppf(0.05, sample_mean, sample_std)
interval_right = norm.ppf(0.95, sample_mean, sample_std)

print(f'Hay un 90% de chances de que el promedio del retorno este en el intervalo ({interval_left}, {interval_right}))')
```
